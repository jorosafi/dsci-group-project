```r
library(tidyverse)
library(repr)
library(tidymodels)
library(readxl)
set.seed(9999)
```

```r
df_default <- read_excel('data/default-cc-data.xls')  %>%
    select(-c(X2:X11))%>%
    rename(client_id = "...1", 
           limit_bal = X1,
           sep_bill = X12 , 
           aug_bill= X13 ,
           july_bill= X14,
           june_bill= X15,
           may_bill= X16, 
           april_bill=X17,
           sep_paid= X18 ,
           aug_paid= X19, 
           july_paid= X20, 
           june_paid= X21 ,
           may_paid= X22 ,
           april_paid= X23,
           default_status = Y)%>%
    slice(2:30001) %>%
    mutate_if(is.character, as.numeric) %>%
    mutate(client_id = as.character(client_id),
          default_status = case_when(default_status =="1"~"yes",
                                     default_status== "0"~"no"),
          default_status = as.factor(default_status)) 
   
    
        
 

head(df_default)
```

```r
default_average = df_default %>%
mutate(average_bill = (sep_bill + aug_bill + july_bill + june_bill + may_bill +april_bill)/6,
      average_paid = (sep_paid + aug_paid + july_paid + june_paid + may_paid +april_paid)/6,
      total_bill = sep_bill + aug_bill + july_bill + june_bill + may_bill +april_bill,
      total_paid = sep_paid + aug_paid + july_paid + june_paid + may_paid +april_paid,
      total_bill_minus_paid = total_bill - total_paid)

head(default_average)
```

```r
options(repr.plot.length = 20)
options(repr.plot.width = 10)
```

```r
sep_bill_plot = ggplot(default_average, aes(x = limit_bal, y = total_bill_minus_paid, colour = default_status)) +
geom_point(alpha = 0.4)+
labs(x = "Remaining Balance", y = "Balance Limit",colour = "default status")
sep_bill_plot
```

```r
tidy_df <- default_average %>%
    pivot_longer(
      cols = sep_bill:april_paid,
      names_to = "bill_payment_date",
      values_to = "bills_payments") %>%
    select(-c(average_bill: total_bill_minus_paid))
head(tidy_df)
```

```r
tidy_df_plot <- tidy_df %>% 
    ggplot(aes(x = bill_payment_date, y = bills_payments, color = default_status)) + 
    geom_point() +

    xlab("month") + 
    ylab("amount") +
    theme(text = element_text(size=20))


tidy_df_plot
```
